# A simple heap

```python
# Plots a min-heap with graphviz.
heap = [1, 3, 5, 9, 4, 7, 8]

import graphviz
dot = graphviz.Digraph(
    node_attr={'shape': 'circle'},
    format='svg',
)
dot.node('R', '1')
dot.edges(['R2', 'R5', '29', '24', '57', '58'])
dot.render('./simple/heap', cleanup=True)  
dot
```

```python
# Step 0: show starting point.
heap = [1, 3, 5, 9, 4, 8, 7]

import graphviz
dot = graphviz.Digraph(
    node_attr={'shape': 'circle'},
    format='svg',
)

dot.node('R', '2')
dot.edges(['R4', 'R5', '49', '47', '58'])

# Invisible last node and edge, so that this graph aligns with the previous one.
dot.node('L', style='invisible')
dot.edge('5', 'L', style='invisible', dir='none')

dot.render('./insertion/00-simple-heap-empty-slot', cleanup=True)
dot
```

```python
# Step 1: highlight next free slot
heap = [1, 3, 5, 9, 4, 8, 7]

import graphviz
dot = graphviz.Digraph(
    node_attr={'shape': 'circle'},
    format='svg',
)

dot.node('R', '2')
dot.edges(['R4', 'R5', '49', '47', '58'])

# Invisible edge so that this graph aligns with the previous one.
dot.node('L', '', style='dashed', color='red')
dot.edge('5', 'L', style='invisible', dir='none')

dot.render('./insertion/01a-highlight-next-free-slot', cleanup=True)
dot
```
